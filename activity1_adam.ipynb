<h1>WHAT TO DO WTH NULL VALUES IN YOUR DATASETS</h1>

<p>THERE ARE TWO WAYS TO HANDLE NULL VALUES</p>
<ol>
    <li><b>Delete the row</b></li>
    <li><b>Fill the empty cell with a common appearing value or average</b></li>
</ol>

```python
birds.isnull().sum()
```

```python
Average_bill_length_mm = birds["bill_length_mm"].mean()
Average_bill_depth_mm = birds["bill_depth_mm"].mean()
Average_flipper_length_mm = birds["flipper_length_mm"].mean()
Average_body_mass_g = birds["body_mass_g"].mean()
MostAppearingSex = birds["sex"].mode()[0] #.mode() always return choices, we must specify to get the top results
Average_weight = birds["weight"].mean()
Average_beak_area = birds["beak_area"].mean()
```

```python
birds["bill_length_mm"] = birds["bill_length_mm"].fillna(Average_bill_length_mm)
birds["bill_depth_mm"] = birds["bill_depth_mm"].fillna(Average_bill_depth_mm)
birds["flipper_length_mm"] = birds["flipper_length_mm"].fillna(Average_flipper_length_mm)
birds["body_mass_g"] = birds["flipper_length_mm"].fillna(Average_body_mass_g)
birds["sex"] = birds["sex"].fillna(MostAppearingSex)
birds["weight"] = birds["weight"].fillna(Average_weight)
birds["beak_area"] = birds["beak_area"].fillna(Average_beak_area)
```

```python
birds.isnull().sum()
```

```python
def getPieChart(column):
    column_array = np.array(column)
    value, occurences = np.unique(column_array, return_counts=True)

    # set the explode effect
    explode = [0.02] #initialize the list
    explode = explode*len(list(occurences)) #sets the explode list according to the length of values
    explode_part = list(occurences).index(occurences.max()) #set where the explode should happen
    explode[explode_part] = 0.08
    return [occurences, value, explode]
```

<h1>DISPLAYING THE DATA ON A PIE CHART</h1>

```python
#initialize canvas and its size
figure, axis = plt.subplots(2, 2)

species_chart = getPieChart(birds["species"])
gender = getPieChart(birds["sex"])
discover = getPieChart(birds["island"])
kailan = getPieChart(birds["year"])

axis[0,0].pie(species_chart[0], labels=species_chart[1], explode=species_chart[2], autopct='%.0f%%')
axis[0,0].set_title("Species Proportion")

axis[0,1].pie(gender[0], labels=gender[1], explode=gender[2], autopct='%.0f%%')
axis[0,1].set_title("Sex Proportion")

axis[1,0].pie(discover[0], labels=discover[1], explode=discover[2], autopct='%.0f%%')
axis[1,0].set_title("Island Discovery Proportion")

axis[1,1].pie(kailan[0], labels=kailan[1], explode=kailan[2], autopct='%.0f%%', textprops = {'color':'#252525'})
axis[1,1].set_title("Island Discovery Proportion")



```
