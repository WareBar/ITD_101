### Importing libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import folium
from folium.plugins import HeatMap

import warnings

warnings.filterwarnings("ignore")
```

### Getting the local datasets

```python
df = pd.read_csv('./datasets/Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df.isnull().sum()
```

## FILTERING DATA FRAME

```python
region_filter_1 = df[df['region_txt'] == "Southeast Asia"] #12.4 K
```

```python
region_filter_2 = region_filter_1[region_filter_1['attacktype1_txt'] == 'Armed Assault'] #1.3K
```

```python
region_filter_3 = region_filter_2[region_filter_2['iyear'] >= 2005] #3K 
```

```python
region_filter_3
```

## DATA AGGREGATION

```python
regional_final_filter = region_filter_3.groupby(['latitude','longitude']).size().reset_index(name='incident_count')
```

```python
regional_final_filter
```

## Converting Pandas Data Frame into Python List

```python
regional_final_filter['normalize_data'] = regional_final_filter['incident_count'] - regional_final_filter['incident_count'].min() / \
                                            regional_final_filter['incident_count'].max() - regional_final_filter['incident_count'].min()
```

```python
regional_heatmap_data = regional_final_filter[['latitude','longitude','normalize_data']].values.tolist()
```

### Creating a base map

```python
regional_map = folium.Map(location = [30, 40], zoom_start=6)
```

### Add Heatmap layer

```python
HeatMap(regional_heatmap_data).add_to(regional_map)
```
